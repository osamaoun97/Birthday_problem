```python
sample_count = 10_000
y = np.array([])
for i in range(1,101):
    Same_bithday_exists_count = 0
    for j in range(sample_count):
        sample = list(np.random.randint(1, high=366,size=i))
        myset = set(sample)
        if len(sample) != len(myset):
            Same_bithday_exists_count = Same_bithday_exists_count + 1
    y = np.append(y, (Same_bithday_exists_count/sample_count))
x = np.arange(1,101)
plt.plot(x,y, 'g')
plt.grid()
plt.xticks(np.arange(0, 101, 10))
plt.xlabel("number of people in the group")
plt.yticks(np.arange(0, 1.1, .1))
plt.ylabel("Probability of 2 people having the same birthday")
plt.title("Probability of at least 2 people sharing the same birthday based on number of people in the group")
```
