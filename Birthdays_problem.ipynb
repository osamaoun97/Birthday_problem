# Birthday problem

https://en.wikipedia.org/wiki/Birthday_problem

**Importing needed libraries**

```python
impoart numpy as np
import matplotlib.pyplot as plt
import math
```

# 1- Theoretical calculations

```python
def birthday_func(no_poeple): # probability of at least two people sharing a birthday versus the number of people
    return 1 - (math.factorial(365)/((365**no_poeple)*math.factorial(365-no_poeple)))

y1 = [birthday_func(i) for i in range(1,101)] # probability for groups from 1 to 100 people
```

# 2- Experimental calculations

```python
sample_count = 10_000
y2 = np.array([])
for i in range(1,101):              # probability for groups from 1 to 100 people
    Same_bithday_exists_count = 0
    for j in range(sample_count):    # calculating probability for each group
        sample = list(np.random.randint(1, high=366,size=i))
        myset = set(sample)
        if len(sample) != len(myset):
            Same_bithday_exists_count = Same_bithday_exists_count + 1
    y2 = np.append(y2, (Same_bithday_exists_count/sample_count))
```

# 3- Plotting the results

```python
x = np.arange(1,101)
fig, axes = plt.subplots(nrows=1, ncols=2,figsize=(12, 5))
plt.suptitle("Probability of at least 2 people sharing the same birthday based on number of people in the group")

axes[0].plot(x,y1, 'g')
axes[0].set_title("Theoretical")
axes[0].set_xlabel("number of people in the group")
axes[0].set_ylabel("Probability of 2 people having the same birthday")
axes[0].grid()
axes[0].set_xticks(np.arange(0, 101, 10))
axes[0].set_yticks(np.arange(0, 1.1, .1))

axes[1].plot(x,y2, 'b')
axes[1].set_title("Experimental")
axes[1].set_xlabel("number of people in the group")
axes[1].set_ylabel("Probability of 2 people having the same birthday")
axes[1].grid()
axes[1].set_xticks(np.arange(0, 101, 10))
axes[1].set_yticks(np.arange(0, 1.1, .1))
plt.show()
```
